```python
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
import matplotlib.gridspec as gridspec
```

```python

fig = plt.figure()
spec = gridspec.GridSpec(ncols=3, nrows=3)
annot_opts = dict(xy=(0.5, 0.5), xycoords='axes fraction',
                 va='center', ha='center')

fig.add_subplot(spec[0, 0]).annotate('GridSpec[0, 0]', **annot_opts)
fig.add_subplot(spec[0, 1:]).annotate('GridSpec[0, 1:]', **annot_opts)
fig.add_subplot(spec[1:, 0]).annotate('GridSpec[1:, 0]', **annot_opts)
fig.add_subplot(spec[1:, 1:]).annotate('GridSpec[1:, 1:]', **annot_opts)

fig.tight_layout()
```
